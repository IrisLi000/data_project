# Scores, Ratings, Gross revenue, Runtime


Boxplot of  data on genres of movies between 1996 and 2020

```python
fig, axes = plt.subplots(2, 2)
fig.set_size_inches(28.7, 22.27)
plt.subplots_adjust(wspace=0.2, hspace=0.4)

sns.boxplot(data=moviedataset_1996_2020_genre1, x='Genre', y='Score', ax=axes[0, 0])
sns.boxplot(data=moviedataset_1996_2020_genre1, x='Genre', y='Rating', ax=axes[0, 1])
sns.boxplot(data=moviedataset_1996_2020_genre1, x='Genre', y='Gross', ax=axes[1, 0])
sns.boxplot(data=moviedataset_1996_2020_genre1, x='Genre', y='RunTime', ax=axes[1, 1])

plt.suptitle('Data according to their genre: 1996-2020')

plt.show()
```

In these boxplots, i am refering to the median, at the minimum and maximum to have a view of the dispersion of the data around the median.

Score: Drama scored the most was crime drama. 

Ratings: Crime drama had the most dispersed ratings raging from 7.5- 9.5, with its median being 8.2-9.0,  obtaining the highest rating out of all the genres. Romance Drama had the 2nd highest rating and dispersion, with its highest being 8.7. 

Gross revenue for movies(millions): Romance drama made the highest gross revenue, followed by war drama. 

Duration of movies: Crime, biography and romance respectively ran the longest duration. While thriller ran the shortest.
